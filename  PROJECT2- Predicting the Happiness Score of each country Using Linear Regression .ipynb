# #Data Cleaning:

```python
df=df.drop(['Standard Error','Family','Trust (Government Corruption)','Generosity'],axis=1)
```

# #Removing Outliers:

Z-Score Technique:

```python
from scipy.stats import zscore
import numpy as np
zsc=zscore(df)
abs_z_score=np.abs(zsc)
filtering_entry=(abs_z_score<3).all(axis=1)
df=df[filtering_entry]
```

# #Model with desired outliers:

```python
y=df['Happiness Score']
x=df.drop(columns=['Happiness Score'])
```

# #Transforming Data to remove skwness:

```python
from sklearn.preprocessing import power_transform
x=power_transform(x)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

# # LinearRegression Model is Used:

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
        
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=43)
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
regression.score(x_test,y_test) #getting regression Score
```

```python
regression.score(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mean_absolute_error(y_test,y_pred) #checking the mean_absolute error
```

```python
mean_squared_error(y_test,y_pred) #checking the mean_square error
```

```python
import numpy as np  #checking the least mean_square error
np.sqrt(mean_squared_error(y_test,y_pred))
```

# #CrossValidation Score for Linear Regression:

```python
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
```

```python
Lassocv=LassoCV(alphas=None,max_iter=1000,normalize=True)
```

```python
Lassocv.fit(x_train,y_train)
```

```python
alpha=Lassocv.alpha_
alpha
```

```python
Lasso_reg=Lasso(alpha)
```

```python
Lasso_reg.fit(x_train,y_train)
```

```python
Lasso_reg.score(x_test,y_test)
```

```python
Ridgecv=RidgeCV(alphas=np.arange(0.001,.1,0.01),normalize=True)
Ridgecv.fit(x_train,y_train)
```

```python
Ridgecv.alpha_
Ridge_reg=Ridge(alpha=Ridgecv.alpha_)
```

```python
Ridge_reg.fit(x_train,y_train)
```

```python
Ridge_reg.score(x_test,y_test)
```
